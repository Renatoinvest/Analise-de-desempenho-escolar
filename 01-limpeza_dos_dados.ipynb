```python
df = pd.read_csv('processed_fechamento_1Bi.csv', skiprows=2, sep=';', encoding='utf-8')
display(df)
```

```python
# Remover colunas "Unnamed"
df = df.loc[:, ~df.columns.str.contains('^Unnamed')]
display(df)
```

```python
# Remover linhas completamente vazias
df.dropna(how='all', inplace=True)
display(df)
```

```python
# Substituir vírgula por ponto e converter para float nas colunas de nota
# Primeiro, identificamos as colunas que não são de nomes
notas_cols = df.columns.drop('Alunos')

for col in notas_cols:
    df[col] = df[col].astype(str).str.replace(',','.').astype(float)
display(df)
```

```python
df.info()
```

```python
# Incluindo a Média final dos alunos

df['Média Final'] = df[notas_cols].mean(axis=1)
display(df)
```

```python
# Arrendondando as casas decimais da média
df['Média Final'] = df['Média Final'].round(1)
display(df)
```

```python
# Ordenar os alunos em ordem alfabética
# 1. Separar alunos reais (nomes que não estão nos níveis de desempenho)
mascaras_finais = ['AVANÇADO', 'ADEQUADO', 'BÁSICO', 'ABAIXO DO BÁSICO']
df_alunos = df[~df['Alunos'].isin(mascaras_finais)].copy()
df_niveis = df[df['Alunos'].isin(mascaras_finais)].copy()

# 2. Ordenar alunos reais em ordem alfabética
df_alunos.sort_values(by='Alunos', inplace=True) 

# 3. Reunir os dois grupos (alunos + níveis de desempenho)
df_final = pd.concat([df_alunos, df_niveis], ignore_index=True)
display(df_final)
```

```python
df_final.to_pickle('df_final.pkl')
```
