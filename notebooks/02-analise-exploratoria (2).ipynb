```python
df_final.to_csv("df_final.csv", index=False)
```

```python
# Carregar o arquivo
df = pd.read_csv("df_final.csv")
```

```python
# Exibir as primeiras linhas
display(df.head())
```

```python
# Verificar colunas de notas (excluindo 'Alunos' e 'Média Final')
colunas_notas = df.columns.drop(['Alunos', 'Média Final'])
```

```python
# Distribuição da Média Final
plt.figure(figsize=(10, 6))
sns.histplot(df['Média Final'], bins=10, kde=True, color='royalblue')
plt.title('Distribuição da Média Final dos Alunos')
plt.xlabel('Média Final')
plt.ylabel('Quantidade de Alunos')
plt.grid(True)
plt.show()
```

```python
# Boxplot por disciplina
plt.figure(figsize=(15, 8))
df_melt = df.melt(id_vars='Alunos', value_vars=colunas_notas, var_name='Disciplina', value_name='Nota')
sns.boxplot(x='Disciplina', y='Nota', data=df_melt)
plt.title('Distribuição das Notas por Disciplina')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Média geral por disciplina
plt.figure(figsize=(12, 6))
media_por_disciplina = df[colunas_notas].mean().sort_values()
sns.barplot(x=media_por_disciplina.values, y=media_por_disciplina.index, palette='viridis')
plt.title('Média Geral por Disciplina')
plt.xlabel('Média das Notas')
plt.tight_layout()
plt.show()
```

```python
# Correlação entre disciplinas
plt.figure(figsize=(12, 8))
corr = df[colunas_notas].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Mapa de Calor das Correlações entre Disciplinas')
plt.tight_layout()
plt.show()
```
