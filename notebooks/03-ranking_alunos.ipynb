## üìå 1. Importar bibliotecas

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Configura√ß√µes visuais
sns.set(style='whitegrid')
plt.rcParams['figure.figsize'] = (10, 6)
```

## üìå 2. Carregar os dados

```python
df_ranking = pd.read_csv('../data/processed/ranking_alunos.csv')
df_ranking.head()
```

```python
# Lista de valores que representam n√≠veis, e n√£o nomes de alunos
niveis_desempenho = ['AVAN√áADO', 'ADEQUADO', 'B√ÅSICO', 'ABAIXO DO B√ÅSICO']

# Filtrar o DataFrame para manter apenas alunos com nomes diferentes dos n√≠veis
df_ranking = df_ranking[~df_ranking['Alunos'].isin(niveis_desempenho)].copy()

# Reiniciar o √≠ndice
df_ranking.reset_index(drop=True, inplace=True)

# Visualizar para confirmar
df_ranking.head()
```

```python
# Lista com os r√≥tulos que queremos remover
mascaras_finais = ['AVAN√áADO', 'ADEQUADO', 'B√ÅSICO', 'ABAIXO DO B√ÅSICO']

# Filtrar somente os alunos (removendo os n√≠veis)
df_alunos = df_ranking[~df_ranking['Alunos'].isin(mascaras_finais)].copy()

# (Opcional) Tamb√©m pode guardar os n√≠veis separadamente, caso queira usar depois:
df_niveis = df_ranking[df_ranking['Alunos'].isin(mascaras_finais)].copy()

# (Opcional) Reordenar por ordem alfab√©tica
df_alunos = df_alunos.sort_values(by='Alunos').reset_index(drop=True)

# Verificar
df_alunos.head()
```

## üìå 3. Calcular a quantidade de notas 10 por aluno

```python
# Selecionar colunas de disciplinas
disciplinas = [
    'F√≠sica', 'Gest√£o de Opera√ß√µes', 'Hist√≥ria',
    'Inova√ß√£o e Des. de Neg√≥cios', 'Projeto Multidisciplinar',
    'Mkt Estrat√©gico', 'Lingua Inglesa', 'Lingua Portuguesa', 'Matem√°tica'
]

# Criar nova coluna com contagem de notas 10
df_ranking['Notas_10'] = df_ranking[disciplinas].apply(lambda x: (x == 10.0).sum(), axis=1)

# Visualizar com a nova coluna
df_ranking[['Alunos', 'Notas_10', 'M√©dia Final']].head()
```

## üìå 4. Gerar Ranking por Quantidade de Notas 10

```python
ranking_notas_10 = df_ranking[['Alunos', 'Notas_10']].sort_values(by='Notas_10', ascending=False).reset_index(drop=True)
ranking_notas_10.head(40)
```

## üìå 5. Gerar Ranking por M√©dia Final

```python
ranking_media = df_ranking[['Alunos', 'M√©dia Final']].sort_values(by='M√©dia Final', ascending=False).reset_index(drop=True)
ranking_media.head(40)
```

## üìå 6. Visualiza√ß√µes com Seaborn - Top 10 por Notas 10

```python
sns.barplot(data=ranking_notas_10.head(10), x='Notas_10', y='Alunos', palette='Blues_d')
plt.title('Top 10 Alunos com Mais Notas 10')
plt.xlabel('Quantidade de Notas 10')
plt.ylabel('Aluno')
plt.tight_layout()
plt.show()
```

## üìå 6. Visualiza√ß√µes com Seaborn - Top 10 por M√©dia Final

```python
sns.barplot(data=ranking_media.head(10), x='M√©dia Final', y='Alunos', palette='Greens_d')
plt.title('Top 10 Alunos com Maior M√©dia Final')
plt.xlabel('M√©dia Final')
plt.ylabel('Aluno')
plt.tight_layout()
plt.show()
```

## üìå 7. (Opcional) Salvar rankings como arquivos CSV

```python
ranking_notas_10.to_csv('../data/processed/ranking_notas_10.csv', index=False)
ranking_media.to_csv('../data/processed/ranking_media_final.csv', index=False)
```
